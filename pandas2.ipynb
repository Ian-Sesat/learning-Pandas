There are different ways to create a dataframe:
1.To read from a csv file
2.To read from an excel file
3.To create a dataframe from a dictionary
4.To create a dataframe from a tuple

1.By reading a csv file as had been earlier read using read_csv

```python
df=pd.read_csv("C:/Users/User/Desktop/Learning ML/Learning Pandas/Learning Data/weather_data_nyc_centralpark_2016.csv")
df
```

2.By reading from an excel file by using the read_excel

```python
df2=pd.read_excel("C:/Users/User/Desktop/Learning ML/Learning Pandas/Learning Data/Robot Arm Budget.xlsx")
df2
```

3.By reading from a dictionary

```python
myDictionary={
    'dates':['17/01/2000','08/01/2000','26/09/2003','08/05/2000','11/09/1999'],
    'rain':[32,33,24,27,20],
    'sun':[25,26,30,32,37],
    'event':["Collo's day","Dottie's day","Ann's Day", "Ian's day","Gyavira's day"]
}
df3=pd.DataFrame(myDictionary)
df3
#df3.columns
```

4.By reading from a tuple

```python
myTuple=[
    ('17/01/2000',32,25,"Collo's day"),
    ('18/01/2000',33,26,"Dottie's day"),
    ('26/09/2003',27,32,"Ann's day"),
    ('11/09/1999',20,37,"Gyavira's day")
]
df4=pd.DataFrame(myTuple,columns=['dates','rain','sun','event'])
df4
```

5.By reading from a list of dictionaries

```python
myDict2=[
    {'date':'17/01/2000','rain':32,'sun':25,'event':"Collo's day"},
    {'date':'08/01/2000','rain':33,'sun':26,'event':"Dottie's day"},
    {'date':'26/09/2003','rain':24,'sun':30,'event':"Ann's day"},
    {'date':'08/05/2000','rain':27,'sun':32,'event':"Ian's day"},
    {'date':'12/09/1999','rain':20,'sun':37,'event':"Collo's day"},
]
df5=pd.DataFrame(myDict2)
df5
```
