```python
import pandas as pd
df1=pd.DataFrame(
{
    'city':['Nakuru','Eldoret','Nairobi'],
    'temperature':[25,20,30],
})
df1
```

```python
df2=pd.DataFrame(
{
    'city':['Eldoret','Nakuru','Nairobi'],
    'humidity':[45,40,30]
})
df2
```

```python
df3=pd.merge(df2,df1,on='city')
df3
```

```python
df4=pd.DataFrame(
{
    'city':['Nakuru','Eldoret','Nairobi','Mombasa'],
    'temperature':[25,20,30,35],
})
df4
```

```python
df5=pd.DataFrame(
{
    'city':['Nakuru','Eldoret','Nyeri','Kisumu'],
    'temperature':[25,20,20,28],
})
df5
```

```python
df6=pd.merge(df5,df4,on='city',how='outer',indicator=True,suffixes=('_left','_right'))
df6
```
