```python
import pandas as pd
import numpy as np
df=pd.read_excel('nyc_weather_data.xlsx')
df
```

```python
newdf=df.replace([-9999,'No effect'],np.nan)
newdf
```

It is better to replace column wise to prevent the  conflict of values

```python
newdf=df.replace({
    'temperature':-9999,
    'windspeed':[-9999,-8888],
    'event':'No event'
})
newdf
```

Use of regular expression to single out the 

```python
newdf=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
newdf
```
