```python
import pandas as pd
df=pd.read_csv('pivot1.csv')
df
```

Suppose I want to arrange the order of my table such that the commomn dates are in the same row, we will use the df.pivot() function

```python
df1=df.pivot(index='date',columns='city')
df1
```

```python
df2=df.pivot(index='city',columns='date')
df2
```

pivot table helps us to summarise and aggregate data inside a dataframe eg when you have a lot of data of a single given day and you want to find its average

```python
df=pd.read_csv('pivot2.csv')
df
```

```python
df1=df.pivot_table(index='date',columns='city',aggfunc='mean',margins=True,margins_name='Average total')
df1
```

Using the incredible pandas we can also find the average rates in a month using the grouper function in the pivot table

```python
df=pd.read_csv('pivot3.csv')
df
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df1=df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
df1
```
