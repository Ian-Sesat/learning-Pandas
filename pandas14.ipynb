```python
import pandas as pd
df=pd.read_csv('Modified1.csv')
df.drop(columns=['Volume'])
```

```python
rng=pd.date_range(start='2022-01-01',end='2022-01-30',freq='B')
rng
```

```python
rng2=pd.date_range(start='2017-01-01',periods=20,freq='B')
rng2
```

```python
df=df.set_index(rng2)
df
```

```python
%matplotlib inline
df.Close.plot()
```

```python
df1=df.asfreq(freq='D',method='ffill')
df1
```

```python
df1.Close.plot()
```

```python
df2=df.asfreq(freq='W',method='pad')
df2
```

```python
df2.Close.plot()
```
